```python
import time
from palettable.matplotlib import matplotlib
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup, SoupStrainer
from geopy.geocoders import Nominatim
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotnine import *
import folium
%matplotlib inline
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf
plt.rc('font', family='NanumGothic')
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')        # Head-less 설정
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=options)
```

```python
def geocoding(address):
    geolocoder = Nominatim(user_agent='South Korea', timeout=None)
    geo = geolocoder.geocode(address)
    crd = {"lat": str(geo.latitude), "lng": str(geo.longitude)}

    return crd
```

```python
pagenum = 2
cnt = 0
driver.get('https://banapresso.com/store')
newName=[]
newSName=[]
newAddress=[]
while(True):
    time.sleep(0.5)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    name = soup.select('span.store_name_map > b')
    s_name = soup.select('span.store_name_map > i')
    address = soup.select('span.store_name_map > span')
    
    for i in range(len(name)):
      eachName = name[i].text
      print(eachName)
      newName.append(eachName)
    for i in range(len(s_name)):
      store_name = s_name[i].text
      print(store_name)
      newSName.append(store_name)
    for i in range(len(address)):
      address1 = address[i].text
      print(address1)
      newAddress.append(address1)
    
    if pagenum == 6:
      driver.find_element('xpath', '//*[@id="contents"]/article/div/section[1]/div/div[1]/div[3]/span/a').click()
      pagenum = 2
      continue
    cnt+=1
    if cnt == 8:
      break
    driver.find_element('xpath', '//*[@id="contents"]/article/div/section[1]/div/div[1]/div[3]/ul/li['+ str(pagenum) +']/a').click()
    pagenum += 1
    
df = pd.DataFrame({ '이름':newName, '점포명':newSName,'주소': newAddress })
df.info()  
    
```

```python
lat = []
lng = []

for index, bana in df.iterrows():
  try:
    crd = geocoding(bana['주소'])
    lat.append(crd['lat'])
    lng.append(crd['lng'])
  except:
    lat.append('')
    lng.append('')
df.loc[:,'lat'] = lat
df.loc[:,'lng'] = lng
idx = df[df['lat'] == '' ].index
print(idx)
df = df.drop(idx)
df['lat'] = df['lat'].astype(float)
df['lng'] = df['lng'].astype(float)

print(df)
bana_map = folium.Map(location=[df['lat'].mean()	,df['lng'].mean()], zoom_start=12)
for index, bana in df.iterrows():
  popup = folium.Popup(bana['점포명'], max_width=100)
  folium.Marker(location=[bana['lat']	, bana['lng']], popup=popup).add_to(bana_map)


bana_map.save('./bana.html')
bana_map
```
